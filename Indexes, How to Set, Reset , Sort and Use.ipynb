```python
import pandas as pd
```

```python
people = {
    "first": ["Ali", "Omar", "Samy"],
    "last": ["Akram", "Ammar", "Ahmed"],
    "email": ["A@d.com", "B@s.com", "C@s.com"]
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.index # Display The Index of DF
```

```python
# Set an Index rather than Default (0, 1, 2, .....)
df.set_index("email", inplace=True)
```

```python
df
```

```python
df.index
```

```python
df.loc["A@d.com"]
```

```python
df.loc["A@d.com", 'last']
```
