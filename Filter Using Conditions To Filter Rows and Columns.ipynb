```python
import pandas as pd
```

```python
people = {
    "first":["Ali", "Ahmed", "Akram"],
    "last":["Ibrahim", "Ayman", "Ammar"],
    "email": ["A@x.com", "B@x.com", "C@x.com"]
}
```

```python
df1 = pd.DataFrame(people)
```

```python
df1
```

```python
df1["last"] == "Ayman"
```

```python
df1[df1["last"] == "Ayman"]
```

```python
filter = df1["last"] == "Ayman"
filter
```

```python
type(filter)
```

```python
df1[df1["last"] == "Ayman"]
```

```python
df1[filter]
```

```python
df1.loc[filter]
```

```python
df1.loc[filter, 'email']
```

```python
# And -> & Or -> | Operators with Conditions

filter = (df1["last"] == "Ammar") & (df1["first"] == "Akram")
```

```python
df1[filter]
```

```python
filter = (df1["last"] == "Ammar") | (df1["first"] == "Akram")
```

```python
df1[filter]
```

```python
# Not -> ~
df1.loc[~filter] # Any Thing that Opposite to the Condition
```
