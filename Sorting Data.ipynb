```python
import pandas as pd 
```

```python
people = {
    "first": ["Coray", "Jhon", "Jane"],
    "last": ["Schafer", "Doe", "Doe"],
    "e-mail": ["CoraySchafer@gmail.com", "JaneDoe@gmail.com", "JohnDoe@gmail.com"]
}
```

```python
df = pd.DataFrame(people)
df
```

```python
# Sort Values
df.sort_values(by='last') # Sort Ascending Alphabetically
```

```python
df.sort_values(by='last', ascending=False) #  Sort Descending Alphabetically
```

```python
df.sort_values(by=['last', 'first'], ascending=False)
# Sort by "last" column , if any duplicates in it -> it will sort by "first" column
```

```python
people = {
    "first": ["Coray", "Jhon", "Jane", "Adam"],
    "last": ["Schafer", "Doe", "Doe", "Doe"],
    "e-mail": ["CoraySchafer@gmail.com", "JaneDoe@gmail.com", "JohnDoe@gmail.com", "AdamDoe@gmail.com"]
}
```

```python
df = pd.DataFrame(people)
```

```python
df.sort_values(by=["last", "first"], ascending=False)
```

```python
df.sort_values(by=["last", "first"], ascending=[False, True])
# Sort by "last" column Descending , if any duplicates in it -> it will sort by "first" column Ascending
```

```python
df.sort_values(by=["last", "first"], ascending=[False, True], inplace=True)
```
