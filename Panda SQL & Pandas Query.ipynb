```python
import pandas as pd
```

```python
students = {"Std_ID":[100, 200, 300, 400],
            "Name": ["Ahmed", "Ali", "Omar", "Yasser"],
            "Mark": [20, 19, 15, 16],
            "Course_ID":[1, 2, 5, 6]}
```

```python
df = pd.DataFrame(students)
```

```python
df
```

```python
df = df.set_index("Std_ID")
```

```python
df
```

```python
# df.query('expression to filter DF', inplace=False)
# inplace: If True, the original DataFrame is modified; otherwise, a new DataFrame is returned.
df.query("Mark > 18")
```

```python
df.query("`Name` == 'Ahmed'")
# Column Name -> Surrounded by `Column Name`
```

```python
df.query("Mark < 20")
```

```python
threshold =  17 # When to use Variable use @
df.query("Mark > @threshold")
```

```python
df["Age"] = [21, 22, 23, 24]
```

```python
df
```

```python
import pandasql as ps

sql_query = "SELECT * FROM df WHERE age > 20"
# DataFrame treated as Database Table

ps.sqldf(sql_query)
```

```python
ps.sqldf(sql_query, locals())
```

```python
sql_query = "SELECT * FROM df WHERE Name LIKE 'A%'"

ps.sqldf(sql_query, locals())
```

```python
sql_query = "SELECT * FROM df WHERE Name LIKE 'A__' "

ps.sqldf(sql_query, locals())
```

```python
sql_query = "SELECT Name, Age FROM df"

ps.sqldf(sql_query, locals())
```

```python
df2 = pd.DataFrame( {"Course_ID":[1, 2, 3, 4, 5, 6],
        "Name": ["DS", "CS", "AI", "Maths", "Algorithms", "C++"] } 
                    )
```

```python
df2
```

```python
sql_query = "SELECT * FROM df d LEFT JOIN df2 d2 ON d.Course_ID = d2.Course_ID"
ps.sqldf(sql_query, locals())
```

```python
sql_query = "SELECT * FROM df d INNER JOIN df2 d2 ON d.Course_ID = d2.Course_ID"
ps.sqldf(sql_query, locals())
```
