### Updating Columns

```python
# To Change Column Name
df.columns = ["First_Name", "Last_Name", "E-mail"]
```

```python
df.columns
```

```python
df
```

```python
# Change Column Name from Lower Case into Upper Case Using List Comprehension
df.columns = [col.upper() for col in df.columns]
```

```python
df
```

```python
# Change Column Name from Upper Case into Lower Case Using List Comprehension
df.columns = [col.lower() for col in df.columns]
```

```python
df
```

```python
# Replace Underscores _ with Spaces Using List Comprehension
df.columns = [col.replace("_", " ") for col in df.columns]
```

```python
df
```

```python
# Replace Spaces with Underscores _  Using List Comprehension 
df.columns = [col.replace(" ", "_") for col in df.columns]
```

```python
df
```

```python
# Anthor Way 
df.columns = df.columns.str.replace("_", " ")
```

```python
df
```

```python
# Rename Column names in a DF
# inplace = False -> Return a New DF
# inplace = True -> Doesnot Return ant thing , but affect the original one
df.rename(columns = {'first name':"FIRST", 'last name':"LAST", "e-mail":"E-MAIL"})
```

```python
df
```

```python
df.rename(columns = {'first name':"FIRST", 'last name':"LAST", "e-mail":"E-MAIL"}, inplace= True)
```

```python
df
```

```python
# If you want to rename specific columns
df.rename(columns = {"FIRST":"first_name"})
```

###  Methods : apply, map, applymap, replace

```python
# apply() : Used to Call Function (Buikt-in or User-Defined) on our Values of DataFrame or Series
df["E-MAIL"].apply(len) # Apply len() Function On Series
```

```python
df["LAST"].apply(len) # Apply len() Function On Series
# We Don't to put () after function
```

```python
def update_email(email):
    return email.upper()
```

```python
df["E-MAIL"].apply(update_email)
```

```python
df
# Doesnot affect the original DF and Returns a new DFdf 
```

```python
df["E-MAIL"] = df["E-MAIL"].apply(update_email)
```

```python
df
```

```python
# You Can apply Lambda (Anonymous Function)
df["FIRST"].apply(lambda name: name.upper())
```

```python
df
```

```python
df["FIRST"] = df["FIRST"].apply(lambda name: name.upper())
```

```python
df
```

```python
# map() --> Works with Series ONLY
# Substitute each Value in Series with Anthor Value
# The Values that we don't substitute it -> It will be converted with NaN

df["FIRST"].map({"ahmed":"cr7", "jane":"messi"})
```

```python
# replace() --> Works with Series ONLY
# Substitute each Value in Series with Anthor Value
# BUTTTT -> The Values that we don't substitute it -> It will not be converted with NaN

df["FIRST"].replace({"ahmed":"cr7", "jane":"messi"})
```
