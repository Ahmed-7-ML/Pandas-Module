```python
df.groupby("Country")["LanguageHaveWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
```

```python
# What Percentage of People from each country know python ?
know_python = df.groupby("Country")["LanguageHaveWorkedWith"].apply(lambda x: x.str.contains("Python").sum())

know_python.sum()
```

```python
(know_python / know_python.sum() )* 100
```

```python
# Anthor Solution
countries = df["Country"].value_counts()
```

```python
python_df = pd.concat([know_python, countries], axis='columns', sort=False) 
python_df
```

```python
python_df.rename(columns = {"LanguageHaveWorkedWith":"NumKnownPython", "Country":"NumRespondents"}, inplace=True)
```

```python
python_df
```

```python
python_df["Percentage"] = python_df["NumKnownPython"] / python_df["NumRespondents"]
```

```python
python_df
```

```python
python_df.sort_values(by='Percentage', ascending=False, inplace=True)
```
